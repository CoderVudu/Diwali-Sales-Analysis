##Exploratory Data Analysis

#Genderwise

```python
ax= sns.countplot(x = 'Gender', data = df, hue = 'Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values('Amount', ascending=False)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values('Amount', ascending=False)
sns.barplot(x= 'Gender', y = 'Amount', data = sales_gen, hue = 'Gender')
plt.show()
```

#Decision: Women have more purchasing power and also purchased more items.

#Age_Group

```python
sns.countplot(x = 'Age Group', data = df, hue = 'Gender')
plt.show()
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values('Amount', ascending=False)
sns.barplot(x= 'Age Group', y = 'Amount', data = sales_age, hue = 'Age Group' )
plt.show()
```

#Decision: This has been found that age gr of 26-35 has more purchasing power and are most frequent buyers as well.

##Statewise

```python
sns.countplot(x = 'State', data = df, hue = 'Orders')
sns.set(rc = {'figure.figsize': (15,5)})
plt.show()
```

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values('Orders', ascending=False)
sns.barplot(x= 'State', y = 'Orders', data = sales_state, hue = 'State')
sns.set(rc = {'figure.figsize': (30,10)})
plt.show()
```

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values('Orders', ascending=False). head(10)
sns.barplot(x= 'State', y = 'Orders', data = sales_state, hue = 'State')
sns.set(rc = {'figure.figsize': (30,10)})
plt.show()
```

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values('Amount', ascending=False). head(10)
sns.barplot(x= 'State', y = 'Amount', data = sales_state, hue = 'State')
sns.set(rc = {'figure.figsize': (30,10)})
plt.show()
```

##We can see that most of the orders are from UP, Maharastra and Karnataka, however Haryana has spent more than Kerala although Kerala did more orders

##Marital Status

```python
sns.set(rc={'figure.figsize': (5, 2)})
ax = sns.countplot(x='Marrige', data=df, hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_Married = df.groupby(['Marrige','Gender'], as_index=False)['Amount'].sum().sort_values('Amount', ascending=False). head(10)
sns.barplot(x= 'Marrige', y = 'Amount', data = sales_Married, hue = 'Gender')
sns.set(rc = {'figure.figsize': (10,5)})
plt.show()
```

#Decision: This has been found that Married Women has more purchasing power and are most frequent buyers as well.

#Occupation_Wise

```python
sns.set(rc={'figure.figsize': (30, 5)})
ax = sns.countplot(x='Occupation', data=df, hue = 'Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_Occ = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values('Amount', ascending=False)
sns.barplot(x= 'Occupation', y = 'Amount', data = sales_Occ, hue = 'Occupation')
sns.set(rc = {'figure.figsize': (30,5)})
plt.show()
```

#We can see that most of the amount spent are from IT, Helathcare and Aviation respectively

#Product_Catagory

```python
sns.set(rc={'figure.figsize': (30, 5)})
ax = sns.countplot(x='Product_Category', data=df, hue = 'Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_P = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values('Amount', ascending=False)
sns.barplot(x= 'Product_Category', y = 'Amount', data = sales_P, hue = 'Product_Category')
sns.set(rc = {'figure.figsize': (30,10)})
plt.show()
```

#We can see that most of the amount spent is from Food

##Product_Id

```python
sales_PI = df.groupby(['Product_ID'], as_index=False)['Amount'].sum().sort_values('Amount', ascending=False).head(10)
sns.barplot(x= 'Product_ID', y = 'Amount', data = sales_PI, hue = 'Product_ID')
sns.set(rc = {'figure.figsize': (30,10)})
plt.show()
```

#CONCLUSION: Married women from age gr 26 to 35 from up, maharastra and karnataka working in the IT, healthcare and Aviation are more likely to buy foods clothing and Electronics items.
